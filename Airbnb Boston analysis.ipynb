## 3. Data Preparation

Clean the data and prepare the data for analysis

```python
# Split data into available units and rented out units
def filter_data(cal, t_or_f):
    '''
    Filter dataframe calendar by (un)availability
    Inputs: calendar dataframe, string
    output: Filtered and cleaned dataframe
    '''
    splitData = cal[cal['available'] == t_or_f]
    splitData['date'] = pd.to_datetime(splitData['date']) + MonthEnd(0)
    return splitData
    
def split_cal(cal, t_or_f,group):
    '''
    Filter dataframe calendar by (un)availability
    Inputs: calendar dataframe, string, group data on key
    output: Filtered, cleaned, and aggregated dataframe
    '''
    splitCal = filter_data(cal, t_or_f)
    splitCal = splitCal.groupby(group).count().reset_index()
    return splitCal
```

```python
def line_plot(df, x, y, name,color, fig=go.Figure()):
    '''
    Plots a line plot
    Inputs: Dataframe to plot, x-axis arg, y-axis arg, name of line, color, figure object
    Output: fig object
    '''
    fig = fig.add_trace(
        go.Scatter(
            x=df[x],
            y=df[y],
            mode='lines+markers',
            line=dict(color=color), 
            name=name))
    return fig
```

```python
# Filter calendar dataframe and clean price column by removing the '$' and converting to float
avg_price = filter_data(cal, 't')
avg_price['price'] = avg_price['price'].apply(lambda x: str(x)[1:])
avg_price['price'] = avg_price['price'].apply(lambda x: x.replace(',','')).astype(float)
# Understand when the average price of units are at it's highest and lowest
avg_price_mean = avg_price.groupby('date').mean().reset_index()
```

### Part 1 - What are the busiest and least busiest times of the year?

Looking at the graph below, we see the number of available and unavailable Airbnb units over the course of a year. So this gives us a look at the busiest and least busiest times of the year. It seems during the holiday season we see the most number of listings on Airbnb, and is when there are more booked units, than available units.

```python
taken = split_cal(cal, 't','date')
free = split_cal(cal, 'f','date')
fig = line_plot(taken, 'date', 'available', 'taken','orange')
fig = line_plot(free,'date','available','free','blue',fig)
fig.update_layout(
    title="# of Listings",
    xaxis_title="Date",
    yaxis_title="# of Listings",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="Blue"
    )
)
fig.show()
```

### Part 2 - Which times of the year are units the most expensive or cheapest?

Looking at this graph, suprisingly the most expensive time of the year isn't during the holiday season. Why is that? We need to break this pricing down further.

### Part 3 - Which neighborhoods are the most expensive?

These next steps we filter and group the data by each neighborhood. The graph is a little cluttered, but we can see some neighborhoods follow different trends for different times of the year. Cambridge and Harvard square are clear cases in which average price goes down during the holiday season, most likely due to college going students heading home for that time, and less people are visiting the campus.

Those two neighborhoods in particular are pretty expensive when it comes to pricing. You can see this in my visual [here](https://public.tableau.com/profile/neil5064#!/vizhome/AirbnbBostonAveragepricingbyNeighborhood2/Dashboard1)
